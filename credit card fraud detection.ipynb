```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to pandas dataframe
credit_card_data = pd.read_csv(r'creditcard.csv')
```

```python
#distribution of legit transaction & fraudulant transaction
credit_card_data['Class'].value_counts()
```

# this dataset is highly unblanced. 0 represents the normal transaction and 1 represents fraudulent transaction


```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape(0))
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

```python
#UNDER SAMPLING
#BUILD A SAMPLE DATASET CONTAINING SIMILAR DISTRIBUTION OF NORMAL TRANSACTION AND FRAUDULANT TRANSACTION
#NUMBER OF FRAUDULANT TRANSACTION I 492

legit_sample = legit.sample(n=492)   #this is for picking the sample randomly
```

```python
legit_sample
```

```python
#CONCATINATING TWO DATAFRAMES

new_dataset = pd.concat([legit_sample, fraud], axis=0)   #axis=0 for adding row wise and axis =1 for adding column wise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting the data into features and targets

X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#split the data into training data and testing data

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
#MODEL TRAINING
#LOGISTIC REGRESSION

model= LogisticRegression()
```

```python
#training the logistic regression model with training data

model.fit(X_train,Y_train)
```

```python
#MODEL EVALUATUION

#ACCURACY SCORE

```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data : ',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Test data : ',test_data_accuracy)
```
